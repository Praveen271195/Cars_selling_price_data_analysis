# Removing Outliers

```python
# Function to extract the list of outliers from the dataset
# IQR = Q3 - Q1
# lower_limit = Q1 - 1.5*IQR
# upper_limit = Q3 - 1.5*IQR

def outliers(dataframe, feature):
    Q1 = dataframe[feature].quantile(0.25)
    Q3 = dataframe[feature].quantile(0.75)
    IQR = Q3-Q1
    lower_limit = Q1 - 1.5*IQR
    upper_limit = Q3 + 1.5*IQR
    
    list_of_indices = dataframe.index[(dataframe[feature] < lower_limit) | (dataframe[feature] > upper_limit)]
    return list_of_indices
```

```python
# Extracting the outliers and storing them in list (will have duplicates)
index_list = [] # Empty set to store output inddices from the multiple columns 
for feature in ["engine_capacity","year","miles_driven","selling_price"]:
    index_list.extend(outliers(cars_data, feature))
```

```python
# Function to remove the set of outliers and return a clean daraframe
def remove_outliers(dataframe, index_list):
    ls = sorted(set(index_list))
    clean_dataframe = dataframe.drop(ls)
    return clean_dataframe
```

```python
# Saving the dataframe as a new version 
cars_data_cleaned = remove_outliers(cars_data, index_list)
cars_data_cleaned.shape
```

## Checking Linearity Again Considering Log Transform of Selling Price

```python
# Final scatter plot combo for visualizing the linearity of each feature with scalled selling price (log transformed)
temp_list = ["miles_driven","engine_capacity","year"]

# Looping through the list of features to be analysed
for i in temp_list:
    sns.regplot(x = i, y = "log_of_selling_price", data = cars_data_cleaned7, line_kws={"color": "red"})
    plt.show()
```

```python
# Dropping the log of miles driven column from the dataset
cars_data_cleaned7.drop("log_of_miles_driven",axis=1,inplace = True)
cars_data_cleaned7.head()
```

```python
# Handling the brand names (categorical variables) by using the concept of Dummy variables
dummy_variables_for_brand = pd.get_dummies(cars_data_cleaned7["vehicle_brand"],drop_first=True)

# Concatenating the dummy variables data with the cars_data
cars_data_cleaned7 = pd.concat([cars_data_cleaned7, dummy_variables_for_brand], axis=1)

# Deleting and reordering the resultant dataframe 
cars_data_cleaned7.drop(["vehicle_brand","selling_price"],axis=1,inplace=True)

# Re-ordering the columns in dataframe
cars_data_cleaned7 = cars_data_cleaned7[['year', 'engine_capacity', 'miles_driven', 'BMW', 'Mercedes-Benz','Mitsubishi','Renault','Toyota','Volkswagen','log_of_selling_price']]
cars_data_cleaned7.head()
```

```python
# Importing a library for splitting the dataset
from sklearn.model_selection import train_test_split

# Input variables
X = cars_data_cleaned7[['year', 'engine_capacity', 'miles_driven', 'BMW', 'Mercedes-Benz',
       'Mitsubishi', 'Renault', 'Toyota', 'Volkswagen']]
# Target variable
y = cars_data_cleaned7["log_of_selling_price"]

# Splitting the dataset into training and testing data; 20% of data will be stored as test data and 80% will be stored
# as train data, for training the linear regression model (for prediction purpose)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# Importing linear model from library called sklearn
from sklearn import linear_model

# Creating the model
regression_model = linear_model.LinearRegression()

# Fitting the model to our train data
regression_model.fit(X_train,y_train)
```

```python
# Prediction results are stored 
y_predictions = regression_model.predict(X_test)
```

```python
# Importing the mean squared error for validation purpose
from sklearn.metrics import mean_squared_error, r2_score, classification_report

# Prints the Coefficients of the model
print("Coefficients: ", regression_model.coef_)
print()

# Intercept
print("Intercept: ", regression_model.intercept_)
print()

# MSE
print("Mean Squared Error: ", mean_squared_error(y_test,y_predictions))
print()

# RMSE:
print("Root Mean Squared Error: ",np.sqrt(mean_squared_error(y_test,y_predictions)))
print()
```

```python
# Overall performace score of the regression model
print("The overall accuracy of the Multiple Linear Regression model: ")
print((r2_score(y_test,y_predictions)*100), "%")
```
