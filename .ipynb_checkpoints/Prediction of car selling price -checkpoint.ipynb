### Checking Linearity

```python
temp_list = ["miles_driven","engine_capacity","year"]

for i in temp_list:
    sns.regplot(x = i, y = "selling_price", data = cars_data, line_kws={"color": "red"})
    plt.show()
```

**Observation:** From the obove 3 graphs, we can see that we can spot some patterns but definitely not linear plots, since we have exponential scatter plots, we can use **Log Transform** on selling price column 

```python
# Current dataframe:
cars_data.head()
```

```python
# Adding new column that contains log transformed values of selling price data which are currenlty too high
# log base 10 will scale the selling price data so that it can be better visualized and liniarity can be observed 
cars_data["log_of_selling_price"] = round(np.log10(cars_data["selling_price"]),4)
cars_data.head()
```

```python
# Adding new column that contains log transformed values of miles_driven data which are currenlty little high
# log base 10 will scale the miles_driven data so that it can be better visualized and liniarity can be observed 
cars_data["log_of_miles_driven"] = round(np.log10(cars_data["miles_driven"]),4)
cars_data.head()
```

```python
# Scatter plot and regression line on log_of_selling_price vs log_of_miles_driven
sns.regplot(x = "log_of_miles_driven", y = "log_of_selling_price", data = cars_data, line_kws={"color": "red"})
plt.show()
```

**Observation:** From the above graph we can see that the linear plot has become worse after taking log transform on the miles driven, now we dont have to use log transformed miled driven data, we will stick with the original unscaled values 

### Checking Linearity Again Considering Log Transform of Selling Price

```python
cars_data.drop("log_of_miles_driven",axis=1,inplace = True)
cars_data
```

```python
# Handling the brand names (categorical variables) by using the concept of Dummy variables
dummy_variables_for_brand = pd.get_dummies(cars_data["vehicle_brand"],drop_first=True)
dummy_variables_for_brand.head()
```

```python
# Concatenating the dummy variables data with the cars_data
cars_data = pd.concat([cars_data, dummy_variables_for_brand], axis=1)
cars_data.head()
```

```python
# Deleting and reordering the resultant dataframe 
cars_data.drop(["vehicle_brand","selling_price"],axis=1,inplace=True)
# Re-ordering the columns in dataframe
cars_data = cars_data[['year', 'engine_capacity', 'miles_driven', 'BMW', 'Mercedes-Benz','Mitsubishi','Renault','Toyota','Volkswagen','log_of_selling_price']]
```

```python
# Importing a library for splitting the dataset
from sklearn.model_selection import train_test_split

# Input variables
X = cars_data[['year', 'engine_capacity', 'miles_driven', 'BMW', 'Mercedes-Benz',
       'Mitsubishi', 'Renault', 'Toyota', 'Volkswagen']]
# Target variable
y = cars_data["log_of_selling_price"]

# Splitting the dataset into training and testing data; 20% of data will be stored as test data and 80% will be stored
# as train data, for training the linear regression model (for prediction purpose)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# Importing linear model from library called sklearn
from sklearn import linear_model

# Creating the model
regression_model = linear_model.LinearRegression()

# Fitting the model to our train data
regression_model.fit(X_train,y_train)
```

```python
# Prediction results are stored 
y_predictions = regression_model.predict(X_test)
```

```python
# Importing the mean squared error for validation purpose
from sklearn.metrics import mean_squared_error, r2_score, classification_report

# Prints the Coefficients of the model
print("Coefficients: ", regression_model.coef_)

# Intercept
print("Intercept: ", regression_model.intercept_)

# MSE
print("Mean Squared Error: ", mean_squared_error(y_test,y_predictions))

# RMSE:
print("Root Mean Squared Error: ",np.sqrt(mean_squared_error(y_test,y_predictions)))
```

```python
# Overall performace score of the regression model
print("The overall accuracy of the Multiple Linear Regression model: ")
print((r2_score(y_test,y_predictions)*100), "%")
```
